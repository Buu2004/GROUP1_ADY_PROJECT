```python
import numpy as np
from sklearn.linear_model import LinearRegression
from numpy import NaN
import matplotlib.pyplot as plt
import pandas as pd
import math 
import os
import glob
```

```python
data_path = 'D:/not sure but it is the Fucking BUGggg!!/DATA FOR ADY/'
```

```python
def read_name_file(link,format):
    txt_files = glob.glob(os.path.join(link,'*.{0}'.format(format)))
    return txt_files
paths = read_name_file(data_path,'csv')
paths
```

```python
titles = []
for path in paths:
    file = path.split('\\')
    tittle = file[-1].replace('.csv','')
    titles.append(tittle)
titles
```

```python
df = []
for i in range(len(paths)):
    df_i = pd.read_csv(paths[i]).T
    df_i.columns = df_i.iloc[0]  # Đặt hàng đầu tiên làm tiêu đề cột
    df_i = df_i[1:]  # Bỏ hàng đầu tiên sau khi đặt tiêu đề cột
    df_i.reset_index(drop=True, inplace=True)  # Đặt lại chỉ số hàng
    df.append(df_i)
merged_data = pd.concat(df, axis=1)
merged_data.index = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
```

```python
merged_data = pd.DataFrame(merged_data)
# Chỉ số dịch vụ
chiso_dv = merged_data['Chỉ số giá sản xuất dịch vụ (Năm trước = 100) chia theo Chỉ số giá sản xuất dịch vụ và Năm'] # lấy cột chỉ số dịch vụ
chiso_dv = pd.DataFrame(chiso_dv)
mean_chiso_dv = chiso_dv.head(11).mean().values # Lấy giá trị mean của 11 hàng đầu tiên

# Mật độ dân số
md_ds = merged_data['Diện tích, dân số và mật độ dân số phân theo địa phương chia theo Địa phương, Năm và Chỉ tiêu']
d_md = md_ds.iloc[11] - md_ds.iloc[10] + md_ds.iloc[11] # Dữ liệu năm 2022 = chỉ số của năm 2021 - năm 2020 + năm 2021 
```

```python
# Sửa đổi dữ liệu cột chỉ số dịch vụ
chiso_dv = chiso_dv.replace({NaN:mean_chiso_dv[0]})

# Sửa đổi dữ liệu mật độ dân số
md_ds = md_ds.replace({NaN:d_md})
```

```python
# Sử lý dữ liệu thiếu của chỉ số dịch vụ
mean_chiso_dv_value = mean_chiso_dv[0]
merged_data['Chỉ số giá sản xuất dịch vụ (Năm trước = 100) chia theo Chỉ số giá sản xuất dịch vụ và Năm'] = chiso_dv.fillna(mean_chiso_dv_value) # Thay giá trị vào merged_data

# Sử lý dữ liệu thiếu của mật độ dân số
merged_data['Diện tích, dân số và mật độ dân số phân theo địa phương chia theo Địa phương, Năm và Chỉ tiêu'] = md_ds.fillna(md_ds)
```
