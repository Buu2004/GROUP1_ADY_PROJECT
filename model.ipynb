```python
def read_name_file(link,format):
    txt_files = glob.glob(os.path.join(link,'*.{0}'.format(format)))
    return txt_files
paths = read_name_file('D:/Hoc Tap/Kỳ 1/PFP191/Python/KY3/PROJECT/','csv')
paths
```

```python
titles = []
for path in paths:
    file = path.split('\\')
    tittle = file[-1].replace('.csv','')
    titles.append(tittle)
titles
```

```python
df = []
for i in range(len(paths)):
    df_i = pd.read_csv(paths[i]).T
    df_i.columns = df_i.iloc[0]  # Đặt hàng đầu tiên làm tiêu đề cột
    df_i = df_i[1:]  # Bỏ hàng đầu tiên sau khi đặt tiêu đề cột
    df_i.reset_index(drop=True, inplace=True)  # Đặt lại chỉ số hàng
    df.append(df_i)
merged_data = pd.concat(df, axis=1)
merged_data.index = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021,2022]
merged_data
```
